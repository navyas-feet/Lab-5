```python
# Here, I have chosen the dataset "iris"  which is a loadable dataset from from Github that seaborn can automatically
#read by using the code %matplotlib inline. I am now loading in the dataset. 

iris = sns.load_dataset("iris")
```

```python
#Here, I am using a boxplot in order to visualize the data and to see where the median, percentiles and outliers lie.
#From this boxplot, we can see that, concerning species and sepal length, there are no outliers expect for one in the 
#species "virginica" and, we can also so that the species "setosa" has the lowest median/range of values while virginica
#has the highest median and the largest range. 

iris = sns.load_dataset("iris")
sns.boxplot(x = "species", y = "sepal_length", data = iris)
```

```python
#Here, we can see that the distribution of petal length based on species varies a lot more than for sepal length. 
#The species setosa has a visibly smaller median and range than the rest of teh species and, it also has more outliers.
#Meanwhile, virginica has the highest median and again, the largest range. Versicolor has one outlier. 

sns.boxplot(x="species", y="petal_length", data=iris)
```

## Data Modelling

```python
#Here, I am modelling the data through the usage of a scatter plot with a regression line to map trends. Here, 
#petal length vs sepal length is being analyzed and, we can determine if there are any relationships beteeen the two 
#through the usage of our linear regression line. 

sns.lmplot(x="petal_length", y="sepal_length", col = "species", data=iris, hue = "species")
```

## Data Analysis

Here, we will be analyzing the scatter plots above in order to see if there is any connection between sepal length and petal length as well comparing the data for both between species. 

As seen above, for the species setosa (blue), it is visible that the values for petal length seem to be concentrated between 1 and 2 inches. From the y-column, we can also see that the values for sepal length are concentrated between 4 and 6 inches. From the line of best fit, we can see that there is a slight positive linear correlation between petal length and sepal length however, it is very slight.

For the species versicolor (orange), it is visible that the values for petal length and sepal length are more distributed than setosa (blue). Versicolor has a larger range for both and, from the graph, we can see that the range of petal length values are from 3 to 5 inches while the range of sepal values are from 5 to 7 inches.  From the line of best fit, we can see that there is a positive linear correlations between petal length and sepal length. 

For the species virginica (green), it is visible that the values for petal length and sepal length are more distributed than both setosa (blue) and versicolor (orange). Versicolor has a larger range for both and, from the graph, we can see that the range of petal length values are from 4.5 to 7 inches while the range of sepal values are from 4 to 8 inches.  From the line of best fit, we can see that there is a positive linear correlations between petal length and sepal length, one that, as seen from the slope, is greater than both versicolor (orange) and setosa (blue). 

## Conclusion

From the visualizations, we can conclude that the species with the highest average petal and sepal length is the virginica. We can also conclude that the species with the lowest average petal and sepal length is the setosa. 

From looking at the lines of best fit, we can also see that the species with the greatest correlation between sepal length and petal length is the virginica, with the setosa having the least correlation between the two. 

The ranges of length values for each of them also vary based on what is seen in the visualizations. As seen in the boxplot under Data Exploration, the species virginica has the largest range and, has the range among the highest values. Meanwhile, setosa has the smallest range among the smallest values. 

Below are sets of histograms to depict the spread and concentration of data for both petal length and sepal length based on species. 

```python
#Here, I am using a facet grid in order to create histograms for each unique value within the column "species". Here, 
#a viewer can easily see the spread and concentration of the data for petal length as well as the maxes, mins and 
#modes. 

grid = sns.FacetGrid(iris, col="species", margin_titles=True)
grid.map(plt.hist, "petal_length");
```

```python
#Here, I am using a facet grid in order to create histograms for each unique value within the column "species". Here, 
#a viewer can easily see the spread and concentration of the data for sepal length as well as the maxes, mins and 
#modes. 

grid = sns.FacetGrid(iris, col="species", margin_titles=True)
grid.map(plt.hist, "sepal_length");
```
